```python
# C:\web_vuln_detector_project\src\predictor.py (và trong Colab, ô Bước 3)

import pandas as pd
import re
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import os
import nltk # Đảm bảo nltk được import

# Định nghĩa lại dataset_base_dir (nếu bạn đã chạy lại kernel)
dataset_base_dir = "csic-2010-dataset"
csv_file_path = os.path.join(dataset_base_dir, "csic_database.csv")

# ----------------------------------------------------------------------
# ĐẢM BẢO ĐÃ MOUNT GOOGLE DRIVE NẾU BẠN SẼ LƯU MÔ HÌNH VÀO ĐÓ SAU NÀY
# (Nếu bạn đã chạy Bước 1 và đã mount, không cần chạy lại)
# from google.colab import drive
# drive.mount('/content/drive')
# ----------------------------------------------------------------------

if not os.path.exists(csv_file_path):
    print(f"Lỗi: Không tìm thấy file CSV tại đường dẫn: {csv_file_path}")
    print("Vui lòng kiểm tra lại tên file và đường dẫn.")
else:
    try:
        df_logs = pd.read_csv(csv_file_path, encoding='latin1', on_bad_lines='skip')
        print(f"Đã tải dữ liệu từ '{csv_file_path}' thành công!")
        print(f"Số lượng bản ghi: {len(df_logs)}")
        print("\nCác cột trong DataFrame:")
        print(df_logs.columns.tolist())

        log_column_for_text = ['Method', 'URL', 'content']
        label_column_name = 'classification'

        missing_log_cols = [col for col in log_column_for_text if col not in df_logs.columns]
        if missing_log_cols:
            raise ValueError(f"Thiếu các cột log cần thiết: {missing_log_cols}. Vui lòng kiểm tra lại tên cột.")
        if label_column_name not in df_logs.columns:
            raise ValueError(f"Thiếu cột nhãn '{label_column_name}'. Vui lòng kiểm tra lại tên cột.")

        df_logs['log_message'] = df_logs['Method'].fillna('') + ' ' + \
                                 df_logs['URL'].fillna('') + ' ' + \
                                 df_logs['content'].fillna('')

        df_logs.rename(columns={label_column_name: 'label'}, inplace=True)

        print("\nDataFrame sau khi tạo cột 'log_message' và đổi tên cột 'label':")
        print(df_logs[['log_message', 'label']].head())
        print("\nGiá trị duy nhất trong cột 'label':")
        print(df_logs['label'].unique())

        # Tiền xử lý văn bản (làm sạch, lemmatization, loại bỏ stopwords)
        lemmatizer = WordNetLemmatizer()
        stop_words = set(stopwords.words('english'))

        # ----------------------------------------------------------------------
        # THAY THẾ HÀM preprocess_text NÀY:
        def preprocess_text(text):
            """
            Hàm tiền xử lý văn bản thay thế:
            - Chuyển đổi về chữ thường
            - Loại bỏ ký tự không phải chữ/số/khoảng trắng
            - Tách từ bằng regex (không dùng nltk.word_tokenize)
            - Lemmatize (đưa về dạng gốc của từ)
            - Loại bỏ stop words (từ phổ biến, ít ý nghĩa như 'a', 'the', 'is')
            """
            if pd.isna(text):
                return ""
            text = str(text).lower()
            text = re.sub(r'[^a-z0-9\s]', '', text)
            # Thay đổi ở đây: Sử dụng re.findall để tokenize thay vì nltk.word_tokenize
            tokens = re.findall(r'\b\w+\b', text) # Tách từ chỉ lấy các từ chữ và số
            tokens = [lemmatizer.lemmatize(word) for word in tokens if word not in stop_words]
            return ' '.join(tokens)
        # ----------------------------------------------------------------------

        df_logs['processed_text'] = df_logs['log_message'].apply(preprocess_text)
        print("\nĐã tiền xử lý văn bản thành công. Dữ liệu sau tiền xử lý:")
        print(df_logs[['log_message', 'processed_text']].head())

        unique_labels = df_logs['label'].unique().tolist()
        unique_labels = [str(l) for l in unique_labels]
        unique_labels.sort()

        label_to_id = {label: i for i, label in enumerate(unique_labels)}
        id_to_label = {i: label for i, label in enumerate(unique_labels)}

        df_logs['label_id'] = df_logs['label'].astype(str).map(label_to_id)
        print(f"\nÁnh xạ nhãn: {label_to_id}")
        print(f"Số lượng lớp (num_labels): {len(unique_labels)}")
        print("\nPhân bố nhãn:")
        print(df_logs['label'].value_counts())

        df_logs.dropna(subset=['processed_text', 'label_id'], inplace=True)
        df_logs['label_id'] = df_logs['label_id'].astype(int)

        if df_logs['label_id'].nunique() < 2:
            print("Cảnh báo: Chỉ tìm thấy một lớp nhãn duy nhất hoặc không đủ lớp sau khi lọc.")
            print("Đảm bảo dataset của bạn có ít nhất hai loại nhãn khác nhau (ví dụ: normal và attack).")
            raise ValueError("Không đủ lớp nhãn để huấn luyện mô hình phân loại. Vui lòng kiểm tra dữ liệu.")

    except Exception as e:
        print(f"Có lỗi xảy ra khi xử lý file CSV: {e}")
        print("Vui lòng kiểm tra lại cấu trúc file CSV và các thông báo lỗi.")
```

# 3. TIỀN XỬ LÝ DỮ LIỆU




```python
# Tải và tiền xử lý dữ liệu
import pandas as pd
import re
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import os
import nltk # Đảm bảo nltk được import
import time # Import time for potential delay

# Định nghĩa lại dataset_base_dir (nếu bạn đã chạy lại kernel)
# Đảm bảo biến này đã được định nghĩa từ bước trước
dataset_base_dir = "csic-2010-dataset"
csv_file_path = os.path.join(dataset_base_dir, "csic_database.csv")

# ----------------------------------------------------------------------
# ĐẢM BẢO ĐÃ MOUNT GOOGLE DRIVE NẾU BẠN SẼ LƯU MÔ HÌNH VÀO ĐÓ SAU NÀY
# (Nếu bạn đã chạy Bước 1 và đã mount, không cần chạy lại)
# from google.colab import drive
# drive.mount('/content/drive')
# ----------------------------------------------------------------------

# Tải các tài nguyên NLTK cần thiết
try:
    nltk.data.find('tokenizers/punkt')
    print("NLTK resource 'punkt' already downloaded.")
except LookupError:
     print("Đang tải tài nguyên NLTK 'punkt'...")
     nltk.download('punkt')

try:
    nltk.data.find('tokenizers/punkt_tab') # Check specifically for punkt_tab
    print("NLTK resource 'punkt_tab' already downloaded.")
except LookupError:
    print("Đang tải tài nguyên NLTK 'punkt_tab'...")
    nltk.download('punkt_tab') # Download punkt_tab explicitly


try:
    nltk.data.find('corpora/stopwords')
    print("NLTK resource 'stopwords' already downloaded.")
except LookupError:
     print("Đang tải tài nguyên NLTK 'stopwords'...")
     nltk.download('stopwords')

try:
    nltk.data.find('corpora/wordnet')
    print("NLTK resource 'wordnet' already downloaded.")
except LookupError:
     print("Đang tải tài nguyên NLTK 'wordnet'...")
     nltk.download('wordnet')

print("Tất cả NLTK resources đã được tải hoặc đã có.")

# Check if the base directory exists before checking for the file
if not os.path.exists(dataset_base_dir):
    print(f"Lỗi: Thư mục dữ liệu '{dataset_base_dir}' không được tìm thấy.")
    print("Vui lòng đảm bảo bạn đã chạy thành công bước tải và giải nén dataset từ Kaggle (thường là ô trước).")
elif not os.path.exists(csv_file_path):
    print(f"Lỗi: Không tìm thấy file CSV tại đường dẫn: {csv_file_path}")
    print(f"Tìm thấy thư mục dữ liệu '{dataset_base_dir}', nhưng file CSV không có trong đó.")
    print("Vui lòng kiểm tra lại tên file trong thư mục dataset.")
else:
    print(f"Kiểm tra lại: File CSV được tìm thấy tại đường dẫn: {csv_file_path}")
    try:
        # Đọc file CSV (encoding latin1 thường hoạt động tốt với các dataset cũ)
        # on_bad_lines='skip' để bỏ qua các dòng lỗi định dạng nếu có
        df_logs = pd.read_csv(csv_file_path, encoding='latin1', on_bad_lines='skip')
        print(f"Đã tải dữ liệu từ '{csv_file_path}' thành công!")
        print(f"Số lượng bản ghi: {len(df_logs)}")
        print("\nCác cột trong DataFrame:")
        print(df_logs.columns.tolist())

        # ----------------------------------------------------------------------
        # ĐIỀU CHỈNH TÊN CỘT ĐỂ PHÙ HỢP VỚI DATASET CỦA BẠN (DỰA TRÊN CSIC 2010)
        # ----------------------------------------------------------------------
        log_column_for_text = ['Method', 'URL', 'content'] # Các cột chứa thông tin request
        label_column_name = 'classification' # Cột chứa nhãn (0 hoặc 1)

        # Kiểm tra sự tồn tại của các cột
        missing_log_cols = [col for col in log_column_for_text if col not in df_logs.columns]
        if missing_log_cols:
            raise ValueError(f"Thiếu các cột log cần thiết: {missing_log_cols}. Vui lòng kiểm tra lại tên cột.")
        if label_column_name not in df_logs.columns:
            raise ValueError(f"Thiếu cột nhãn '{label_column_name}'. Vui lòng kiểm tra lại tên cột.")

        # Tạo cột 'log_message' bằng cách kết hợp các cột liên quan
        # Thêm khoảng trắng để phân tách rõ ràng các phần của request
        df_logs['log_message'] = df_logs['Method'].fillna('') + ' ' + \
                                 df_logs['URL'].fillna('') + ' ' + \
                                 df_logs['content'].fillna('')

        # Đổi tên cột nhãn để chuẩn hóa cho Hugging Face Trainer
        df_logs.rename(columns={label_column_name: 'label'}, inplace=True)

        print("\nDataFrame sau khi tạo cột 'log_message' và đổi tên cột 'label':")
        print(df_logs[['log_message', 'label']].head())
        print("\nGiá trị duy nhất trong cột 'label':")
        print(df_logs['label'].unique()) # Sẽ là [0 1] hoặc ['normal', 'attack_type']

        # Tiền xử lý văn bản (làm sạch, tokenization, lemmatization, loại bỏ stopwords)
        lemmatizer = WordNetLemmatizer()
        stop_words = set(stopwords.words('english'))

        def preprocess_text(text):
            """
            Hàm tiền xử lý văn bản:
            - Chuyển đổi về chữ thường
            - Loại bỏ ký tự không phải chữ/số/khoảng trắng
            - Tách từ (tokenize)
            - Lemmatize (đưa về dạng gốc của từ)
            - Loại bỏ stop words (từ phổ biến, ít ý nghĩa như 'a', 'the', 'is')
            """
            if pd.isna(text): # Xử lý NaN nếu có
                return ""
            text = str(text).lower() # Chuyển tất cả sang chữ thường
            text = re.sub(r'[^a-z0-9\s]', '', text) # Loại bỏ ký tự không phải chữ/số/khoảng trắng
            tokens = nltk.word_tokenize(text) # Tách từ
            tokens = [lemmatizer.lemmatize(word) for word in tokens if word not in stop_words] # Lemmatize và loại bỏ stopwords
            return ' '.join(tokens)

        df_logs['processed_text'] = df_logs['log_message'].apply(preprocess_text)
        print("\nĐã tiền xử lý văn bản thành công. Dữ liệu sau tiền xử lý:")
        print(df_logs[['log_message', 'processed_text']].head())

        # Ánh xạ nhãn chuỗi sang ID số (nếu chưa phải số)
        # Dataset CSIC 2010 đã có nhãn 0 và 1, nên việc ánh xạ sẽ đơn giản
        unique_labels = df_logs['label'].unique().tolist()
        # Đảm bảo các nhãn là chuỗi nếu không, để tránh lỗi JSON dump sau này
        unique_labels = [str(l) for l in unique_labels] # Chuyển đổi nhãn về dạng chuỗi
        unique_labels.sort() # Sắp xếp để đảm bảo ánh xạ consistent (ví dụ 0->Normal, 1->Attack)

        label_to_id = {label: i for i, label in enumerate(unique_labels)}
        id_to_label = {i: label for i, label in enumerate(unique_labels)}

        df_logs['label_id'] = df_logs['label'].astype(str).map(label_to_id) # Đảm bảo map đúng
        print(f"\nÁnh xạ nhãn: {label_to_id}")
        print(f"Số lượng lớp (num_labels): {len(unique_labels)}")
        print("\nPhân bố nhãn:")
        print(df_logs['label'].value_counts())

        # Lọc bỏ các hàng có nhãn NaN hoặc không hợp lệ (quan trọng cho stratify trong train_test_split)
        df_logs.dropna(subset=['processed_text', 'label_id'], inplace=True)
        # Đảm bảo label_id là kiểu số nguyên
        df_logs['label_id'] = df_logs['label_id'].astype(int)

        if df_logs['label_id'].nunique() < 2:
            print("Cảnh báo: Chỉ tìm thấy một lớp nhãn duy nhất hoặc không đủ lớp sau khi lọc.")
            print("Đảm bảo dataset của bạn có ít nhất hai loại nhãn khác nhau (ví dụ: normal và attack).")
            # Nếu chỉ có 1 nhãn, không thể huấn luyện phân loại
            raise ValueError("Không đủ lớp nhãn để huấn luyện mô hình phân loại. Vui lòng kiểm tra dữ liệu.")


    except Exception as e:
        print(f"Có lỗi xảy ra khi xử lý file CSV: {e}")
        print("Vui lòng kiểm tra lại cấu trúc file CSV và các thông báo lỗi.")
```
